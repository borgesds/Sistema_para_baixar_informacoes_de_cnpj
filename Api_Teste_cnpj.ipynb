## M√©todo # 1: usando a compreens√£o de lista

Essa √© uma das maneiras pelas quais essa tarefa pode ser executada. Nela, iteramos para cada dicion√°rio dentro da lista e verificamos se h√° uma chave nela, se estiver presente, o valor exigido √© retornado.

```python
test_list = [{"Gfg" : 3, "b" : 7},  
             {"is" : 5, 'a' : 10},  
             {"Best" : 9, 'c' : 11}] 
print("The original list : " + str(test_list)) 
K = 'Best'
res = [sub[K] for sub in test_list if K in sub][0] 
print("The extracted value : " + str(res)) 
```

## M√©todo 2: Usando update() + loop

Esta √© mais uma maneira pela qual essa tarefa pode ser realizada. Nisto, atualizamos cada dicion√°rio um no outro. Formando um grande dicion√°rio, o valor √© extra√≠do deste dicion√°rio.

```python
test_list = [{"Gfg" : 3, "b" : 7},  
             {"is" : 5, 'a' : 10},  
             {"Best" : 9, 'c' : 11}] 
print("The original list : " + str(test_list)) 
K = 'Best'

res = dict() 
for sub in test_list: 
      
    res.update(sub) 
print("The extracted value : " + str(res[K])) 
```
